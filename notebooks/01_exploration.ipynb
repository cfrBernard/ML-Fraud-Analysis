```python
# %%
import seaborn as sns
import matplotlib.pyplot as plt

# Graph 1 : distribution des montants (transactions normales)
plt.figure(figsize=(10, 4))
sns.histplot(df[df["Class"] == 0]["Amount"], bins=100, log_scale=(False, True))
plt.title("Montant des transactions normales (log scale Y)")
plt.show()

# Graph 2 : distribution des montants (fraudes)
plt.figure(figsize=(10, 4))
sns.histplot(df[df["Class"] == 1]["Amount"], bins=100, log_scale=(False, True), color="red")
plt.title("Montant des fraudes (log scale Y)")
plt.show()
```

```python
# %%
correlation_matrix = df.corr(numeric_only=True)
correlation_matrix
```

```python
# %%
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 12))
sns.heatmap(correlation_matrix, cmap="coolwarm", center=0, annot=False)
plt.title("Matrice de corrÃ©lation")
plt.show()
```

```python
# %%
correlation_with_class = correlation_matrix["Class"].sort_values(ascending=False)
correlation_with_class
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x='Class', y='V11', data=df)
plt.title('Distribution de V11 selon Class')
plt.show()
```
