```python
# %%
import seaborn as sns
import matplotlib.pyplot as plt

# Graph 1 : distribution des montants (transactions normales)
plt.figure(figsize=(10, 4))
sns.histplot(df[df["Class"] == 0]["Amount"], bins=100, log_scale=(False, True))
plt.title("Montant des transactions normales (log scale Y)")
plt.show()

# Graph 2 : distribution des montants (fraudes)
plt.figure(figsize=(10, 4))
sns.histplot(df[df["Class"] == 1]["Amount"], bins=100, log_scale=(False, True), color="red")
plt.title("Montant des fraudes (log scale Y)")
plt.show()
```
