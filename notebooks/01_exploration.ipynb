```python
import numpy as np
import pandas as pd

def scan_fraud_zones(df, variables, bins=10, min_rate=0.01, max_rate=0.15, min_count=30):
    results = {}
    for var in variables:
        vals = df[var].dropna()
        min_val, max_val = vals.min(), vals.max()
        bins_edges = np.linspace(min_val, max_val, bins + 1)

        suspicious_zones = []

        for i in range(bins):
            low, high = bins_edges[i], bins_edges[i+1]
            segment = df[(df[var] >= low) & (df[var] < high)]
            count = len(segment)
            if count < min_count:
                continue
            fraud_rate = segment['Class'].mean()
            if min_rate <= fraud_rate <= max_rate:
                suspicious_zones.append({
                    'range': (low, high),
                    'fraud_rate': fraud_rate,
                    'count': count
                })

        if suspicious_zones:
            results[var] = suspicious_zones

    return results


# Exemple
variables_of_interest = ['V4', 'V10', 'V11', 'V12', 'V14', 'V17']
zones = scan_fraud_zones(df, variables_of_interest, bins=20, min_rate=0.70, max_rate=0.90, min_count=20)

for var, zones_list in zones.items():
    print(f"\nVariable: {var}")
    for zone in zones_list:
        print(f"  Range: {zone['range']}, Fraud rate: {zone['fraud_rate']*100:.2f}%, Count: {zone['count']}")
```

```python
# bins feature
def compute_risk_features(row, risk_dict):
    score = 0
    risk_counts = {"low": 0, "mid": 0, "high": 0}
    
    for level, features in risk_dict.items():
        level_score = {"low": 1, "mid": 2, "high": 3}[level]
        for feature, bins in features.items():
            value = row.get(feature, None)
            if value is None:
                continue
            for b in bins:
                low, high = b["range"]
                if low <= value < high:
                    score += level_score
                    risk_counts[level] += 1
                    break

    return pd.Series({
        "risk_score": score,
        "risk_low": risk_counts["low"],
        "risk_mid": risk_counts["mid"],
        "risk_high": risk_counts["high"]
    })
```

```python
# 1. Calcul des features de risque (bins)
risk_features = df.apply(compute_risk_features, axis=1, risk_dict=risk_dict)

# 2. Flag de zone de montant Ã  risque
df['amount_high_fraud_zone'] = df['Amount'].apply(amount_risk_flag)

# 3. Merge
df = pd.concat([df, risk_features], axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Liste des colonnes Ã  visualiser
features = [col for col in df.columns if col not in ["Time", "Class"]]

for feature in features:
    plt.figure(figsize=(14, 6))

    # KDE Plot
    plt.subplot(1, 2, 1)
    sns.kdeplot(df[df["Class"] == 0][feature], label="Non-Fraud", fill=True)
    sns.kdeplot(df[df["Class"] == 1][feature], label="Fraud", fill=True, color="red")
    plt.title(f"KDE Plot - {feature}")
    plt.legend()

    # Box Plot
    plt.subplot(1, 2, 2)
    sns.boxplot(x="Class", y=feature, data=df)
    plt.title(f"Boxplot - {feature}")

    plt.tight_layout()
    plt.show()
```
